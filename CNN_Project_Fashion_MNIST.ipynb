# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import keras
```

# Load Data

```python
# it already provide data in train and test based
(X_train,y_train),(X_test,y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
class_labels = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
'''
0 => T-shirt/top
1 => Trouser
2 => Pullover
3 => Dress
4 => Coat
5 => Sandal
6 => Shirt
7 => Sneaker
8 => Bag
9 => Ankle boot'''
```

#Change Dimension

beacause CNN take 4d dimension

```python
X_train.ndim
```

```python
X_train.shape
```

```python
X_train = np.expand_dims(X_train,-1)
X_test = np.expand_dims(X_test,-1)
```

```python
X_train.ndim
```

```python
X_train.shape
```

# Feature Scaling

```python
X_train = X_train/255
X_test = X_test/255
```

# Split Dataset
from traing data to train and validation data because CNN required validation data at training time

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(X_train,y_train,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_val.shape
```

# Convolutional Neural Network - Model Building

```python
model = keras.models.Sequential([
              keras.layers.Conv2D(filters=32,kernel_size=3,strides=1,padding="valid",activation="relu",input_shape=[28,28,1]),
              keras.layers.MaxPool2D(pool_size=(2,2),),
              keras.layers.Flatten(),
              keras.layers.Dense(units=128,activation="relu"),
              keras.layers.Dense(units=10,activation="softmax") # softmax provide how much probability of that product
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=512,verbose=1,validation_data=(X_val,y_val))
```

# Test the Model

```python
# model.predict(X_test[0]) -> it will give error because x_test is not in 4 dimension
model.predict(np.expand_dims(X_test[0],axis=0)).round(2)
```

```python
np.argmax(model.predict(np.expand_dims(X_test[0],axis=0)).round(2))
```

```python
y_test[0]
```

```python
y_pred = model.predict(X_test).round(2)
```

```python
y_pred
```

```python
model.evaluate(X_test,y_test)
```

# Visualize Output

```python
plt.figure(figsize=(16,30))
j = 1
for i in np.random.randint(0,1000,60):
  plt.subplot(10,6,j)
  j+=1
  plt.imshow(X_test[i].reshape(28,28), cmap="Grays")
  plt.title('Actual = {} / {}\nPredicted = {} / {}'.format(class_labels[y_test[i]],y_test[i],class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
  plt.axis("off")
```

# Confustion Matrix

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(16,9))
y_pred_levels = [np.argmax(label) for label in y_pred]
cm = confusion_matrix(y_test,y_pred_levels)

#show cm
sns.heatmap(cm,annot=True,fmt="d",xticklabels=class_labels,yticklabels=class_labels)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_pred_levels,target_names=class_labels)
print(cr)
```
